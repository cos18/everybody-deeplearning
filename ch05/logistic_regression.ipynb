### 예제 : 로지스틱 회귀
#### 공부한 시간과 합격 여부 사이 관계 분석하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2, 0], [4, 0], [6, 0], [7, 1], [9, 0], [10, 1], [12, 1], [14, 1]]

time_data = [d[0] for d in data]
pass_data = [d[1] for d in data]

plt.scatter(time_data, pass_data)
plt.xlim(0, 15)
plt.ylim(-.1, 1.1)
```

### 학습률에 따른 예측값 변화

```python
a = 0
b = 0
learning_rate = 0.02

def sigmoid(x):
    return 1 / (1 + np.e ** (-x))

for i in range(10001):
    for t_data, p_data in data:
        a_diff = t_data * (sigmoid(a * t_data + b) - p_data)
        b_diff = sigmoid(a * t_data + b) - p_data
        a = a - learning_rate * a_diff
        b = b - learning_rate * b_diff
    if i % 1000 == 0:
        print("epoch={} a={} b={}".format(i, a, b))
        
t_range = np.arange(0, 15, 0.1)
plt.scatter(time_data, pass_data)
plt.plot(t_range, np.array([sigmoid(a * t + b) for t in t_range]))
plt.show()
```
