### 예제 : 평균 제곱 오차
#### 공부한 시간과 성적 사이 관계 분석하기 (2)

```python
import numpy as np
import matplotlib.pyplot as plt

time = [1, 2, 4, 6, 8, 10, 12]
score = [79, 87, 84, 91, 97, 93, 100]

plt.figure(figsize=(12, 6))
plt.scatter(time, score)
plt.show()
```

```python
# 초기 설정
fake_a = 3.7
fake_b = 70

predict = [fake_a * t + fake_b for t in time]

plt.scatter(time, score)
plt.plot([min(time), max(time)], [min(predict), max(predict)])
plt.show()
```

$$MSE=\frac {1}{n}\sum \left( \widehat {y}_{i}-y_{i}\right) ^{2}$$

```python
div = sum([(p - s) ** 2 for p, s in zip(predict, score)])
mse = div / len(score)
print(mse)
```
